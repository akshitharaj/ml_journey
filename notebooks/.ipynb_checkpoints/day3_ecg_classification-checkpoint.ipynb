```python
#load dataset
import pandas as pd

df = pd.read_csv("../datasets/ecg/mitbih_train.csv")
df.head()
```

```python
df.shape
df.iloc[:,-1].value_counts()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
#normalize
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42)
```

```python
#baseline model : logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred_rf)

plt.figure()
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
#classification report
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf))
```

```python
#plot sample ECG signal
import matplotlib.pyplot as plt

plt.plot(X.iloc[0])
plt.title("Sample ECG Signal")
plt.show()
```

```python
#observations:
# - Dataset contains ECG heartbeat signals
# - Baseline ML accuracy: 91% accuracy
# - Random Forest performed better with 97% accuracy
# - ECG signals are sequential → CNN likely better
# - Class 0 (normal beats) has high recall due to dataset dominance.
# - Class 1 shows the lowest recall, indicating difficulty in detecting this arrhythmia type.
# - Minority classes suffer from misclassification due to class imbalance.
# - Accuracy alone is not sufficient for evaluating medical classification models.
# - Future improvement should focus on increasing recall for abnormal classes.
```

```python
#Clinical Interpretation:

# - In ECG arrhythmia detection, recall for abnormal classes is more important than precision. 
# - Missing an abnormal heartbeat (false negative) could have serious clinical consequences. 
# - Although class 4 shows high precision, its low recall indicates that many abnormal beats are not being detected.
# - Future improvements should focus on increasing recall for minority arrhythmia classes.
```
