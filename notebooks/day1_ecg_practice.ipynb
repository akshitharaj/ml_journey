```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500)

ecg = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 15 * t)

noise = 0.3 * np.random.randn(len(t))
noisy_ecg = ecg + noise

window_size = 5
filtered_ecg = np.convolve(noisy_ecg, 
                            np.ones(window_size)/window_size, 
                            mode='same')

plt.figure(figsize=(10,5))
plt.plot(t, noisy_ecg, label="Noisy ECG", alpha=0.5)
plt.plot(t, filtered_ecg, label="Filtered ECG", linewidth=2)
plt.legend()
plt.title("ECG Signal Filtering Practice")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 500)

ecg = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 15 * t)

noise = 0.3 * np.random.randn(len(t))
noisy_ecg = ecg + noise

window_size = 15
filtered_ecg = np.convolve(noisy_ecg, 
                            np.ones(window_size)/window_size, 
                            mode='same')

plt.figure(figsize=(10,5))
plt.plot(t, noisy_ecg, label="Noisy ECG", alpha=0.5)
plt.plot(t, filtered_ecg, label="Filtered ECG", linewidth=2)
plt.legend()
plt.title("ECG Signal Filtering Practice")
plt.show()
```

```python
# Observations:
# 1. Noise makes ECG harder to interpret.
# 2. Moving average smooths signal.
# 3. Larger window = smoother but more distortion.
```
