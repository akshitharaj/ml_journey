```python
# -- Basic syntax:
# -- [expression for item in iterable]
#create list of squares
numbers = [1, 2, 3, 4, 5]

squares = [x**2 for x in numbers]

print(squares)
```

```python
# with condition
numbers = [1,2,3,4,5,6]

even_numbers = [x for x in numbers if x % 2 == 0]

print(even_numbers)
```

```python
# Dictionary
# -- A dictionary stores key–value pairs.
student = {
    "name": "Alice",
    "age": 20,
    "marks": 85
}

print(student["name"])
```

```python
# loop through dictionary
for key, value in student.items():
    print(key, value)
```

```python
# Lambda Functions
# -- A lambda is a small anonymous function.
# -- Basic syntax:
# -- lambda arguments: expression
add = lambda a, b: a + b

print(add(3,5))
```

```python
square = lambda x: x**2

print(square(6))
```

```python
# Sorting with key
# -- Python’s sorted() or .sort() can use a key function.
numbers = [-10, 5, -3, 8]

sorted_numbers = sorted(numbers, key=abs)

print(sorted_numbers)
```

```python
# Example exercise
students = [
    {"name": "Alice", "marks": 85},
    {"name": "Bob", "marks": 72},
    {"name": "Charlie", "marks": 90},
    {"name": "David", "marks": 78}
]

students_sorted = sorted(students, key=lambda x: x["marks"])

print(students_sorted)
```

```python
# sort highest marks first
students_sorted = sorted(students, key=lambda x: x["marks"], reverse=True)

print(students_sorted)
```

```python
#challenge
students_1 = [
    {"name": "Tom", "marks": 65},
    {"name": "Anna", "marks": 92},
    {"name": "John", "marks": 75},
    {"name": "Lily", "marks": 88}
]

sort_marks = sorted(students_1, key=lambda x: x["marks"])
print(sort_marks)
```

```python
#list of student names
names = [x["name"] for x in students_1]
print(names)
```

```python
#name marks mapping
name_marks = {student["name"]: student["marks"] for student in students_1}

print(name_marks)
```
