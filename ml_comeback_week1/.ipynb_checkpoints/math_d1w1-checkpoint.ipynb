```python
import numpy as np
import statistics
import pandas as pd

data = [10, 20, 30, 40, 50]
data_df = pd.DataFrame(data)

# --- Numpy method
mean_np = np.mean(data)
median_np = np.median(data)
std_np = np.std(data) # Population std dev
var_np = np.var(data) # Population variance

# --- Statistics method
mean_st = statistics.mean(data)
median_st = statistics.median(data)
std_st = statistics.stdev(data) # Sample std dev
var_st = statistics.variance(data) # Sample variance

# --- pandas method
mean_pd = data_df.mean()
median_pd = data_df.median()
std_pd = data_df.std()
var_pd = data_df.var()
```
