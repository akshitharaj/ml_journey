```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create time axis
t = np.linspace(0, 1, 500)
#this represents one second of signal
```

```python
#create ecg waves
p_wave = 0.1 * np.sin(2 * np.pi * 5 * t)
qrs = np.exp(-((t-0.3)/0.02)**2) * 1.5
t_wave = 0.3 * np.exp(-((t-0.6)/0.05)**2)

ecg = p_wave + qrs + t_wave
```

```python
#plot the signal
plt.figure(figsize=(10,4))

plt.plot(t, ecg)

plt.title("Simulated ECG Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")

plt.grid()
plt.show()
```
