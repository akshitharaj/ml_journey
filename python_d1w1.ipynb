```python
#list() constructor
onelist = list(("apple", "banana", "chikkoo"))
print(onelist)
```

```python
#if 
if "chikkoo" in onelist:
    print("yes 'chikkoo' is in the list")
```

```python
onelist[2] = "strawberry"
print(onelist)
onelist[1:2] = "orange", "blueberry"
print(onelist)
```

```python
#insert
onelist.insert(5, "watermelon")
print(onelist)
```

```python
#append
onelist.append("grapes")
print(onelist)
```

```python
twolist = ["kiwi", "papaya", "mango"]
onelist.extend(twolist)
print(onelist)
```

```python
#looping
thirdlist = ["apple", "banana", "chikkoo"]
for i in thirdlist:
    print(i)
```

```python
for i in range(len(thirdlist)):
    print(thirdlist[0])
```

```python
#while loop
i = 0
while i < len(thirdlist):
    print(thirdlist[i])
    i = i+1
```

```python
#looping using list comprehension
[print(x) for x in thirdlist] 
```

```python
newlist = []
for x in thirdlist:
    if 'a' in x:
        newlist.append(x)

print(newlist)
```

```python
new = [x for x in thirdlist if 'a' in x]
print(new)
```
